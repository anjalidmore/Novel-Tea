```python
# Select Important Columns

movies = df[['id','title','overview','genres','keywords']]
movies.head()
```

```python
# Convert Genres to List

import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
```

```python
# Convert Keywords

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Convert Overview Into Word List

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Remove Spaces in Multi-word Genres

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Create Tags Column

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']
```

```python
# Convert Tags to String

movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
# Final Dataset for Recommendation

new_df = movies[['id','title','tags']]
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer # CountVectorizer converts text into word frequency vectors.
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english') # Vector Initialization
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # convert tags into vectors
```

```python
# Apply Stemming 
# Stemming reduces words to their root form. (here. using NLTK)

# Example:
# loved → love  
# loving → love

import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```
